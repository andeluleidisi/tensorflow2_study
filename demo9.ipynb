```python
import  numpy as np
import matplotlib.pyplot as plt
import mpl_toolkits.axisartist as axisartist
```

```python
# sigmoid函数导数
def sigmoid(x):
    '''
    公式：1/(1+exp(-x))
    :param x:
    :return:
    '''
    return 1/(1+np.exp(-x))

def derivative(x):
    '''
    :param x:
    :return: sigmoid函数的导数
    '''
    return sigmoid(x)*(1-sigmoid(x))

x = np.array(range(-10,11))
print(x)
plt.figure()
plt.plot(x,sigmoid(x),x,derivative(x))
plt.show()
```

```python
def relu(x):
    x[x<0] = 0
    print(x)
    return x
def derivative(x):
    d = np.array(x)
    d[x<0] = 0
    d[x>=0] = 1
    return d
x = np.array(range(-10,11))
print(x)
plt.figure()
plt.plot(x,relu(x),x,derivative(x))
plt.show()
```

```python
def leaky_relu(x):
    d = np.array(x,dtype=float)
    d[x<0] = 0.2*d[x<0]
    return d
def derivative(x):
    d = np.array(x,dtype=float)
    d[x<0] = 0.2
    d[x>=0] = 1
    return d
x = np.linspace(-10,10,1000)
plt.figure()
plt.plot(x,relu(x),x,derivative(x))
plt.grid()
plt.show()
```

```python
# tanh
def tanh(x):
    x = np.array(x,dtype=float)
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
def derivative(x):
    return 1-tanh(x)**2

x = np.linspace(-10,10,100)
plt.figure()
plt.plot(x,tanh(x),x,derivative(x))
plt.grid()
plt.show()
```
